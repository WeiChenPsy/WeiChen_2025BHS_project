```python
import os
import numpy as np
import pandas as pd
import nibabel as nib
from nilearn.image import concat_imgs, math_img
from nilearn.input_data import NiftiMasker

def extract_and_save_roi_ts(subject_list, roi_paths, base_dir, output_dir, save_rest_img=True):
    os.makedirs(output_dir, exist_ok=True)
    rest_img_dir = os.path.join(output_dir, "rest")
    if save_rest_img:
        os.makedirs(rest_img_dir, exist_ok=True)

    for subject in subject_list:
        print(f"\n處理中：{subject}")
        mni_dir = os.path.join(base_dir, subject, "MNI")

        # 合併 4 個 run 的影像
        run_imgs = []
        for run in range(1, 5):
            fname = f"{subject}_task-rest_run-{run}_bold.nii.gz"
            fpath = os.path.join(mni_dir, fname)
            if os.path.exists(fpath):
                run_imgs.append(fpath)
            else:
                print(f"找不到檔案：{fpath}")
        if not run_imgs:
            print("未找到任何影像，略過此受試者。")
            continue
        concat_img = concat_imgs(run_imgs)
        print(f"成功合併 {len(run_imgs)} 個 run")

        # 儲存合併後的 resting-state 影像
        if save_rest_img:
            rest_img_path = os.path.join(rest_img_dir, f"{subject}_task-rest_concat.nii.gz")
            concat_img.to_filename(rest_img_path)
            print(f"已儲存合併後影像：{rest_img_path}")

        # 對每個 ROI 處理
        for roi_idx, roi_path in enumerate(roi_paths, start=1):
            roi_id = f"roi{roi_idx}"
            print(f"  處理 ROI：{roi_id}")
            roi_img = nib.load(roi_path)
            roi_bin = math_img("img > 0", img=roi_img)

            masker = NiftiMasker(mask_img=roi_bin, standardize=True)
            roi_ts = masker.fit_transform(concat_img)
            mean_ts = roi_ts.mean(axis=1)

            # 儲存時間序列
            npy_path = os.path.join(output_dir, f"{subject}_{roi_id}_ts.npy")
            csv_path = os.path.join(output_dir, f"{subject}_{roi_id}_ts.csv")
            np.save(npy_path, mean_ts)
            pd.DataFrame({f"{roi_id}_mean_signal": mean_ts}).to_csv(csv_path, index=False)

            print(f"已儲存時間序列：")
            print(f" NPY：{npy_path}")
            print(f" CSV：{csv_path}")
```

```python
### extract time series from RSN & analysis the correlation to ROI time series 
import os
import numpy as np
import pandas as pd
from nilearn.input_data import NiftiMasker

# 路徑設定
subjects = [f"sub-{i:02d}" for i in range(1, 7)]
roi_dir = "output/roi_masks"
rsn_dir = "output/rsn_masks_resampled"
rest_dir = "output/roi_ts/rest"
roi_ts_dir = "output/roi_ts"
output_csv = "output/roi_rsn_fc_results.csv"

# 載入 ROI 與 RSN mask 檔案
roi_files = [
    "output/roi_masks/group_roi_1.nii.gz",
    "output/roi_masks/group_roi_2.nii.gz"
]

rsn_files = {
    "DMN": os.path.join(rsn_dir, "DMN_mask_resampled.nii.gz"),
    "FPN": os.path.join(rsn_dir, "FPN_mask_resampled.nii.gz"),
    "SN":  os.path.join(rsn_dir, "SN_mask_resampled.nii.gz"),
}

results = []

for sub in subjects:
    print(f"開始處理 {sub}")
    bold_path = os.path.join(rest_dir, f"{sub}_task-rest_concat.nii.gz")
    if not os.path.exists(bold_path):
        print(f"找不到影像：{bold_path}，略過此受試者")
        continue

    for roi_idx, roi_path in enumerate(roi_files, start=1):
        roi_id = f"roi{roi_idx}"
        print(f"  處理 ROI：{roi_id}")
        csv_path = os.path.join(roi_ts_dir, f"{sub}_{roi_id}_ts.csv")

        if not os.path.exists(csv_path):
            print(f"    找不到 ROI 時間序列：{csv_path}，略過此 ROI")
            continue

        ts_roi_mean = pd.read_csv(csv_path).values.squeeze()
        row = {"subject": sub, "roi": roi_id}

        for rsn_name, rsn_path in rsn_files.items():
            print(f"    計算 ROI 與 {rsn_name} 的相關性")
            masker_rsn = NiftiMasker(mask_img=rsn_path, standardize=True)
            ts_rsn = masker_rsn.fit_transform(bold_path)
            ts_rsn_mean = ts_rsn.mean(axis=1)

            r = np.corrcoef(ts_roi_mean, ts_rsn_mean)[0, 1]
            row[f"{rsn_name}_r"] = r

        results.append(row)

# 輸出結果
df = pd.DataFrame(results)
df.to_csv(output_csv, index=False)
print(f" ROI–RSN functional connectivity：{output_csv}")
```

### extract time series from RSN & ROI (z score-sample standardize)

```python
import os
import numpy as np
from nilearn.input_data import NiftiMasker
from nilearn.image import load_img, math_img

# === 路徑設定 ===
subjects = [f"sub-{i:02d}" for i in range(1, 7)]
roi_dir = "output/roi_masks"
rsn_dir = "output/rsn_masks_resampled"
rest_dir = "output/roi_ts/rest"
roi_ts_dir = "output/roi_ts_zscore"
rsn_ts_dir = "output/rsn_ts_zscore"
os.makedirs(roi_ts_dir, exist_ok=True)
os.makedirs(rsn_ts_dir, exist_ok=True)

# === ROI 與 RSN 檔案 ===
roi_files = {
    "roi1": os.path.join(roi_dir, "group_roi_1.nii.gz"),
    "roi2": os.path.join(roi_dir, "group_roi_2.nii.gz"),
}
rsn_files = {
    "DMN": os.path.join(rsn_dir, "DMN_mask_resampled.nii.gz"),
    "FPN": os.path.join(rsn_dir, "FPN_mask_resampled.nii.gz"),
    "SN":  os.path.join(rsn_dir, "SN_mask_resampled.nii.gz"),
}

# === 提取並儲存時序的共用函式 ===
def extract_and_save_ts(subject, label, mask_path, output_dir, is_roi=True):
    rest_path = os.path.join(rest_dir, f"{subject}_task-rest_concat.nii.gz")
    if not os.path.exists(rest_path):
        print(f"檔案不存在：{rest_path}")
        return

    mask_img = load_img(mask_path)
    mask_bin = math_img("img > 0", img=mask_img)

    masker = NiftiMasker(mask_img=mask_bin, standardize="zscore_sample")
    ts = masker.fit_transform(rest_path).mean(axis=1)

    base_filename = f"{subject}_{label}_zscore_ts"
    npy_path = os.path.join(output_dir, base_filename + ".npy")
    csv_path = os.path.join(output_dir, base_filename + ".csv")
    np.save(npy_path, ts)
    np.savetxt(csv_path, ts, delimiter=",")

    print(f"儲存完成：{subject} - {label} → {csv_path} {npy_path}")

# === 處理 ROI 時序 ===
for sub in subjects:
    for roi_label, roi_path in roi_files.items():
        print(f"ROI｜處理 {sub} - {roi_label}")
        extract_and_save_ts(sub, roi_label, roi_path, roi_ts_dir, is_roi=True)

# === 處理 RSN 時序 ===
for sub in subjects:
    for rsn_label, rsn_path in rsn_files.items():
        print(f"RSN｜處理 {sub} - {rsn_label}")
        extract_and_save_ts(sub, rsn_label, rsn_path, rsn_ts_dir, is_roi=False)
```

```python
### 防止 NAN 被丟棄
import os
import numpy as np
from nilearn.input_data import NiftiMasker
from nilearn.image import load_img, math_img

# === 路徑設定 ===
subjects = [f"sub-{i:02d}" for i in range(1, 7)]
roi_dir = "output/roi_masks"
rest_dir = "output/roi_ts/rest"
roi_ts_dir = "output/roi_ts_zscore_nan"
os.makedirs(roi_ts_dir, exist_ok=True)

# === ROI 檔案 ===
roi_files = {
    "roi1": os.path.join(roi_dir, "group_roi_1.nii.gz"),
    "roi2": os.path.join(roi_dir, "group_roi_2.nii.gz"),
}

# === 提取並儲存 ROI 時序 ===
def extract_roi_ts_with_nanmean(subject, roi_label, mask_path):
    rest_path = os.path.join(rest_dir, f"{subject}_task-rest_concat.nii.gz")
    if not os.path.exists(rest_path):
        print(f"❌ 檔案不存在：{rest_path}")
        return

    mask_img = load_img(mask_path)
    mask_bin = math_img("img > 0", img=mask_img)

    masker = NiftiMasker(mask_img=mask_bin, standardize="zscore_sample")
    ts = masker.fit_transform(rest_path)  # shape = (time_points, n_voxels)

    mean_ts = np.nanmean(ts, axis=1)  # ❗防止因 NaN 丟失時間點

    # 儲存檔案
    base_filename = f"{subject}_{roi_label}_zscore_ts"
    npy_path = os.path.join(roi_ts_dir, base_filename + ".npy")
    csv_path = os.path.join(roi_ts_dir, base_filename + ".csv")
    np.save(npy_path, mean_ts)
    np.savetxt(csv_path, mean_ts, delimiter=",")

    print(f"✅ 儲存完成：{csv_path}（{len(mean_ts)} 時間點）")

# === 執行全部受試者與 ROI ===
for sub in subjects:
    for roi_label, roi_path in roi_files.items():
        print(f"處理 {sub} - {roi_label}")
        extract_roi_ts_with_nanmean(sub, roi_label, roi_path)
```
