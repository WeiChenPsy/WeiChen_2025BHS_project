```python
from scipy.io import loadmat

mat = loadmat("raw_data/ds004078/derivatives/annotations/time_align/word-level/story_1_word_time.mat")

# 查看有哪些 key
print(mat.keys())
```

```python
import pandas as pd

onsets = mat["start"].flatten()
offsets = mat["end"].flatten()
durations = offsets - onsets
words = [w[0] for w in mat["word"].flatten()]

df = pd.DataFrame({
    "onset": onsets,
    "offset": offsets,
    "duration": durations,
    "word": words
})
```

```python
df[df["duration"] == 0]
```

```python
print("事件總數：", len(df))
print("duration 為 0 的事件數：", (df["duration"] == 0).sum())
```

```python
import nibabel as nib
from utils import get_paths

subject = "sub-01"
n_runs = 60  # 或你目前的 run 數

run_lengths = []

for run in range(1, n_runs + 1):
    func_file, _ = get_paths(subject, run)
    img = nib.load(func_file)
    n_volumes = img.shape[3]
    run_lengths.append(n_volumes)
    print(f"Run {run:02d} has {n_volumes} volumes.")

# 總和核對（確認是否為 masked_data 的總時間點數）
print("Total timepoints:", sum(run_lengths))
```
