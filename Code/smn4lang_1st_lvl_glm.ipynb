```python
from nilearn.input_data import NiftiMasker
from nilearn.image import concat_imgs
from nilearn.glm.first_level import run_glm
from nilearn.glm.first_level import make_first_level_design_matrix
from nilearn.glm.contrasts import compute_contrast
from scipy.io import loadmat
import nibabel as nib
import numpy as np
import pandas as pd
import os
```

```python
# setting
subject = "sub-01"
n_runs = 2  # 這裡只用前 2 個 run
tr = 0.71
```

```python
### events
from utils import get_paths

# 儲存影像檔路徑和事件資訊
func_imgs = []
all_events = []
offset = 0  # 記錄每個 run 的時間起點

for run in range(1, n_runs + 1):
    func_file, mat_file = get_paths(subject, run)
    func_imgs.append(func_file)  # 不先讀進記憶體，交給 masker 處理

    mat = loadmat(mat_file)
    onsets = mat["start"].flatten()
    offsets = mat["end"].flatten()
    durations = offsets - onsets

    # 篩掉 duration = 0 的事件
    valid = durations > 0
    onsets = onsets[valid] + offset
    durations = durations[valid]

    events = pd.DataFrame({
        "onset": onsets,
        "duration": durations,
        "trial_type": "speech"
    })
    all_events.append(events)

    # 加總時間偏移量
    img = nib.load(func_file)
    offset += img.shape[-1] * tr  # 每個 run 的時間長度（n_scans * tr）

# 結合所有事件資訊
all_events_df = pd.concat(all_events, ignore_index=True)
```

```python
### Masker
# 設定 masker（自動產生 mask、標準化、smoothing）
masker = NiftiMasker(
    standardize=True,
    smoothing_fwhm=6,
    memory='nilearn_cache',
    verbose=1
)
```

```python
### Combine img
from nilearn.image import concat_imgs

# 將多個 4D 影像合併成一個大 4D 影像（time concat）
concat_img = concat_imgs(func_imgs)

import os
_ = os.system('say -v Kyoko "なになに!!!"')
```

```python
# 套用 masker（這一步會花時間）
masked_data = masker.fit_transform(concat_img)  # shape = (T_total, N_voxels)

# 建立設計矩陣
n_scans = masked_data.shape[0]
frame_times = np.arange(n_scans) * tr

design_matrix = make_first_level_design_matrix(
    frame_times,
    all_events_df,
    hrf_model='spm',
    drift_model='cosine'
)
# 擬合 GLM（使用 run_glm）
labels, results = run_glm(masked_data, design_matrix.values)

import os
_ = os.system('say -v Kyoko "全然わかんない"')
```

```python
# 指定對比（"speech" 對 baseline）
contrast = np.array([1 if name == "speech" else 0 for name in design_matrix.columns])
z_map_data = compute_contrast(labels, results, contrast, stat_type='t').z_score()
z_map_img = masker.inverse_transform(z_map_data)
```

```python
# 還原為 NIfTI z-map
z_map_img = masker.inverse_transform(z_map_data)
```

```python
from nilearn import plotting

# 載入 anatomical template（或你自己的 T1 影像也可以）
from nilearn.datasets import load_mni152_template
anat_img = load_mni152_template()

# 顯示圖像
display = plotting.plot_stat_map(
    z_map_img,
    bg_img=anat_img,
    threshold=3.1,
    display_mode='z',
    cut_coords=[-40, -25, -10, 0, 10, 25, 40],
    title=f"{subject} Speech > Baseline",
    black_bg=True
)
```

```python
### threshold
from nilearn.glm import threshold_stats_img
z_map_thresh, _ = threshold_stats_img(z_map_img, 
                                      height_control='fdr', 
                                      alpha=0.001, 
                                      cluster_threshold=30)
```

```python
# 顯示圖像
display = plotting.plot_stat_map(
    z_map_thresh,
    bg_img=anat_img,
    threshold=3.1,
    display_mode='z',
    cut_coords=[-40, -25, -10, 0, 10, 25, 40],
    title=f"{subject} Speech > Baseline",
    black_bg=True
)
```
