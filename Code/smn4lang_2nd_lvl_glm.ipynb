```python
from nilearn.image import concat_imgs, mean_img
from nilearn.glm.second_level import SecondLevelModel
from nilearn.plotting import plot_stat_map
import pandas as pd
import os
```

```python
# 受試者清單
subjects = ['sub-01', 'sub-02', 'sub-03', 'sub-04', 'sub-05', 'sub-06']

# 影像路徑
zmap_paths = [f"output/1stlvl/{sub}_zmap.nii.gz" for sub in subjects]
```

```python
# 建立 second-level input dataframe
design_matrix = pd.DataFrame({'intercept': [1] * len(subjects)})

# 看一看
from IPython.display import display
display(design_matrix)
```

```python
# 擬合 second-level model
second_level_model = SecondLevelModel()
second_level_model = second_level_model.fit(zmap_paths, design_matrix=design_matrix)

# 對全體做 one-sample t-test
z_map = second_level_model.compute_contrast(second_level_contrast='intercept', output_type='z_score')
```

```python
# 顯示圖像
plot_stat_map(z_map, 
              threshold=3.1, 
              display_mode='z', 
              cut_coords=[-40, -25, -10, 0, 10, 25, 40],
              title='Group-level Speech Activation',
              black_bg=True
             )
```

```python
from nilearn.glm import threshold_stats_img

z_map_thresh, threshold = threshold_stats_img(
    z_map,
    alpha=0.05,
    height_control='fdr',
    cluster_threshold=200,
    two_sided=True)

print(f"Threshold: {threshold}")
```

```python
from nilearn import datasets

from nilearn.reporting import get_clusters_table
table = get_clusters_table(
    z_map_thresh,
    stat_threshold=threshold,  
    cluster_threshold=10,
    two_sided=True
)
table
```

```python
# Peak MNI 座標
peak_coords = table[['X', 'Y', 'Z']].values
```

```python
from nilearn import datasets
import nibabel as nib
import numpy as np

# 載入 Harvard-Oxford atlas
atlas = datasets.fetch_atlas_harvard_oxford('cort-maxprob-thr25-2mm')
atlas_img = atlas.maps
atlas_data = atlas_img.get_fdata()
atlas_affine = atlas_img.affine
labels = atlas.labels
```

```python
from nilearn.image import coord_transform

atlas_labels = []

for coord in peak_coords:
    # MNI → voxel 座標
    ijk = np.round(nib.affines.apply_affine(np.linalg.inv(atlas_affine), coord)).astype(int)

    try:
        label_idx = int(atlas_data[ijk[0], ijk[1], ijk[2]])
        atlas_labels.append(labels[label_idx] if label_idx < len(labels) else 'Unknown')
    except:
        atlas_labels.append('Out of bounds or error')
```

```python
table['Atlas Label'] = atlas_labels
table
```

```python
from nilearn.plotting import view_img

# 顯示互動式的 z-map（你也可以換成 z_map_thresh）
interactive_view = view_img(z_map, threshold=3.1, title='Group Speech Activation')

# 顯示視圖
interactive_view
```

```python
# 儲存 z-map 圖像
z_map.to_filename("output/2ndlvl/group_speech_zmap.nii.gz")
z_map_thresh.to_filename("output/2ndlvl/group_speech_zmapthresh.nii.gz")

# saved table
table.to_csv("output/2ndlvl/group_cluster_table.csv", index=False)

# 也儲存圖片
from nilearn import plotting
display = plotting.plot_stat_map(
    z_map_thresh,
    threshold=3.1,
    display_mode='z',
    cut_coords=[-40, -25, -10, 0, 10, 25, 40],
    title='Group-level Speech Activation (alpha = 0.05)',
    black_bg=True
)
display.savefig("output/2ndlvl/group_speech_zmap_thresh.png")
```

ROI

```python
from nilearn.regions import connected_regions

# 萃取 threshold 過後的 ROI
regions_img, index = connected_regions(z_map_thresh, min_region_size=200)

# 查看總共幾個 ROI
print(f"總共找到 {len(index)} 個 ROI")
```

```python
import os
from nilearn.image import index_img

output_dir = "output/roi_masks"
os.makedirs(output_dir, exist_ok=True)

for i in range(regions_img.shape[-1]):
    roi_i = index_img(regions_img, i)
    roi_path = os.path.join(output_dir, f"group_roi_{i+1}.nii.gz")
    roi_i.to_filename(roi_path)
    print(f"Saved：{roi_path}")
```

```python
from nilearn.plotting import plot_roi
from nilearn import datasets

bg_img = datasets.load_mni152_template()
plot_roi(roi_i, bg_img=bg_img, display_mode='z', cut_coords=[-40, -25, -10, 0, 10, 25, 40])
```
